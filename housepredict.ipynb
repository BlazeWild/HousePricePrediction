PREDICTION MODEL

```python
X = data.drop(columns = ["price"])
y = data["price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.compose import ColumnTransformer
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

APPLY LINEAR REGRESSION

```python
columns_trans = make_column_transformer((OneHotEncoder(sparse=False), ['mainroad', 'furnishingstatus'])
                                        , remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
#lr = LinearRegression(normalize = True)
```

```python
pipe = make_pipeline(columns_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

***Applying Lasso***

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(columns_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test, y_pred_lasso)
```

***Applying Ridge***

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(columns_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
Pipeline(steps = [('columntransformer', ColumnTransformer(remainder = 'passthrough', 
                transformers=[('onehotencoder', OneHotEncoder(sparse=False),['furnishingstatus'])])),
                  ('standardscaler', StandardScaler()), ('ridge', Ridge())])
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization: ", r2_score(y_test, y_pred_lr ))
print("Lasso: ", r2_score(y_test, y_pred_lasso))
print("Ridge: " , r2_score(y_test, y_pred_ridge))
```
